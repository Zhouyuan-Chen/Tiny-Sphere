```python
import meshplot
import scipy as sp
import numpy as np
import igl
import meshplot as mp
# import open3d as o3d
```

```python
meshplot.jupyter()
v, f = igl.read_triangle_mesh("../data/cat.off")
# pi /= 10
n = igl.per_vertex_normals(v, f)
mp.plot(v, shading={"point_size": 8})
# v.shape
# ni.shape
```

```python

import matplotlib.pyplot as plt

pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(v)
pcd.normals = o3d.utility.Vector3dVector(n)

poisson_mesh, densities = o3d.geometry.TriangleMesh.create_from_point_cloud_poisson(pcd, depth=9)

# æå–æ ‡é‡åœº (densities)
vertices = np.asarray(poisson_mesh.vertices)
densities = np.asarray(densities)

# å¯è§†åŒ–æ ‡é‡åœº
plt.scatter(vertices[:, 0], vertices[:, 1], c=densities, cmap='viridis')
plt.colorbar(label='Density')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scalar Field from Poisson Reconstruction')
plt.show()
```
